```python
from specutils import Spectrum1D, SpectralRegion, Spectrum
from specutils.manipulation import extract_region
from astrodb_utils.spectra import check_spectrum_plottable
from astropy import units as u
import matplotlib.pyplot as plt
```

```python
path = "/Users/guanying/SIMPLE db/SIMPLE-db/scripts/spectra_convert/zhang18/sty2054_supplemental_files"
```

```python
import os
print(os.path.exists("/Users/guanying/SIMPLE Archieve/SIMPLE-db/scripts/ingests/Zhang18/sty2054_supplemental_files/SDSS_J134749.74+333601.7_sdL0_SDSS_Primeval-I.txt"))
```

This notebook is to convert OSIRIS instrument data:     
- 64 OSIRIS file

```python

file_plotted = 0
file_failed = 0
failed_files = []

for filename in os.listdir(path):
    if filename.endswith(".fits") and "osiris" in filename.lower():
        file_path = os.path.join(path, filename)
        try:
            print(f"Reading OSIRIS FITS file: {filename}")

            # Load spectrum
            spectrum = Spectrum.read(file_path)

            # Convert flux units properly
            spectrum = spectrum.with_flux_unit(
                u.erg / (u.s * u.cm**2 * u.AA),
                equivalencies=u.spectral_density(spectrum.spectral_axis)
            )

            # Extract spectral region
            if filename == "ULAS_J230443.30+093423.9_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits":
                print("Before extraction:")
                check_spectrum_plottable(spectrum, show_plot=True)
                print("Extracting region 0.625-0.98")
                spectrum = extract_region(spectrum, SpectralRegion(6250*u.AA, 9800*u.AA))
            elif filename == "ULAS_J141832.35+025323.0_sdL0_comb_OSIRIS_scombine_Primeval-IV.fits":
                print("Before extraction:")
                check_spectrum_plottable(spectrum, show_plot=True)
                print("Extracting region 0.62-0.975")
                spectrum = extract_region(spectrum, SpectralRegion(6200*u.AA, 9750*u.AA))
            # else:
            #     print("Before extraction:")
            #     check_spectrum_plottable(spectrum, show_plot=True)
            #     print("Extracting region 0.54-1.02")
            #     spectrum = extract_region(spectrum, SpectralRegion(5400*u.AA, 10200*u.AA))
                
            # Check and plot
            if check_spectrum_plottable(spectrum, raise_error=True, show_plot=True):
                print(f"Plottable file: {filename}\n")
                file_plotted += 1

        except Exception as e:
            print(f"Error reading {filename}: {e}")
            file_failed += 1
            failed_files.append(filename)
            continue

print(f"\nTotal files plotted: {file_plotted}")
print(f"Total files failed: {file_failed}")
print("Failed files: ", failed_files)
```
