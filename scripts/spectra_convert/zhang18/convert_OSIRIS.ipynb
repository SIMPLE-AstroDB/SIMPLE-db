```python
def make_new_fitsfiles(path, output_path):
    file_plotted = 0
    file_failed = 0
    failed_files = []

    os.makedirs(output_path, exist_ok=True)

    for filename in os.listdir(path):
        if "OSIRIS" not in filename:
            continue

        file_path = os.path.join(path, filename)
        try:
            print(f"\nReading OSIRIS FITS file: {filename}")

            # Open FITS and fix units if needed
            with fits.open(file_path, mode='update') as hdul:
                header = hdul[0].header
                data = hdul[0].data
                print(f"  Original BUNIT: {header.get('BUNIT', 'N/A')}")

                header.set("BUNIT", "erg / (cm2 s Angstrom)", comment="Corrected from original BUNIT")
                print(f"    header BUNIT corrected to: {header['BUNIT']}")

            # Create Spectrum
            spectral_axis = np.arange(len(data)) * u.nm
            flux = data * u.erg / (u.cm**2 * u.s * u.AA)
            spectrum = Spectrum(spectral_axis=spectral_axis, flux=flux)

            # Determine masking for specific files
            mask = None
            low = high = None


            if filename in [
                "ULAS_J230443.30+093423.9_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J231924.35+052524.5_esdL1_comb_OSIRIS_scombine_Primeval-IV.fits",
                "ULAS_J001931.33+063111.0_M9_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J141203.85+121609.9_sdL5_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J141832.35+025323.0_sdL0_comb_OSIRIS_scombine_Primeval-IV.fits",
                "ULAS_J231949.36+044559.5_M7_comb_OSIRIS_scombine_Primeval-IV.fits",
                "ULAS_J021258.08+064115.9_sdL1_OSIRIS_GTC39-12B_Primeval-IV.fits"
                ]: low, high = 300, 1000
                

            # Masking
            if low is not None and high is not None:
                print(f"  Masking region {low}-{high}")


                print("    Original spectrum")
                check_spectrum_plottable(spectrum,show_plot=True)

                # Apply mask 
                mask = np.logical_or(
                    spectrum.spectral_axis.value < low,
                    spectrum.spectral_axis.value > high
                )

                masked_spectrum = Spectrum(
                    spectral_axis=spectrum.spectral_axis,
                    flux=spectrum.flux,
                    mask=mask
                )

                # Plot masked spectrum 
                print("    Masked spectrum")
                masked_spectrum.plot()
                fig, ax = plt.subplots()
                ax.plot(masked_spectrum.spectral_axis[~masked_spectrum.mask], masked_spectrum.flux[~masked_spectrum.mask], color='orange', label='Unmasked Region')
                ax.plot(masked_spectrum.spectral_axis[masked_spectrum.mask], masked_spectrum.flux[masked_spectrum.mask], color='blue', label='Masked Region')
                ax.set_xlabel("Wavelength")
                ax.set_ylabel("Flux (erg / cm / s / AA)")
                ax.legend()
                plt.show()
                file_plotted += 1

                # Add mask to header
                header.set("MASK_LOW", low, comment="Lower bound of masked region")
                header.set("MASK_HIGH", high, comment="Upper bound of masked region")

            else:
                # Just check plottable and plot once
                if check_spectrum_plottable(spectrum, raise_error=True, show_plot=True):
                    print(f"  Plottable file: {filename}")
                    file_plotted += 1

            # Save processed spectrum
            out_filename = filename.replace(".fits", "_SIMPLE.fits")
            out_file = os.path.join(output_path, out_filename)
            fits.writeto(out_file, spectrum.flux.value, header=header, overwrite=True)
            print(f"  Saved processed spectrum: {out_filename}")

        except Exception as e:
            print(f"  ERROR reading {filename}: {e}")
            file_failed += 1
            failed_files.append(filename)
            continue

    print(f"\nTotal files plotted: {file_plotted}")
    print(f"Total files failed: {file_failed}")
    if failed_files:
        print("Failed files: ", failed_files)


# Run
make_new_fitsfiles(path, output_path)
```
