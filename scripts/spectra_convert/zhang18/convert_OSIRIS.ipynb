```python
def make_new_fitsfiles(path, output_path):
    file_plotted = 0
    file_failed = 0
    failed_files = []

    os.makedirs(output_path, exist_ok=True)

    for filename in os.listdir(path):
        if "OSIRIS" not in filename:
            continue

        file_path = os.path.join(path, filename)
        try:
            print(f"\nReading OSIRIS FITS file: {filename}")

            # Open FITS and fix units if needed
            with fits.open(file_path, mode='update') as hdul:
                header = hdul[0].header.copy()
                data = hdul[0].data
                print(f"  Original BUNIT: {header.get('BUNIT', 'N/A')}")

                header.set("BUNIT", "erg / (cm2 s Angstrom)", comment="Corrected from original BUNIT")
                print(f"    header BUNIT corrected to: {header['BUNIT']}")

            # Create Spectrum
            spectral_axis = np.arange(len(data)) * u.nm
            flux = data * u.erg / (u.cm**2 * u.s * u.AA)
            spectrum = Spectrum(spectral_axis=spectral_axis, flux=flux, meta={'header': header})

            # Determine masking for specific files
            mask = None
            low = high = None

            if filename in [
                "ULAS_J231924.35+052524.5_esdL1_comb_OSIRIS_scombine_Primeval-IV.fits",
                "ULAS_J141203.85+121609.9_sdL5_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J231949.36+044559.5_M7_comb_OSIRIS_scombine_Primeval-IV.fits",
                "ULAS_J021258.08+064115.9_sdL1_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J231924.35+052524.5_esdL1_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J231924.35+052524.5_esdL1_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J082206.61+044101.8_sdL0.5_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J143154.18-004114.3_sdM9_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J233227.03+123452.0_M6_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J231924.35+052524.5_esdL1_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J223123.44+010025.1_M6_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J151649.84+083607.1_sdL5_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J125938.50+301500.2_M9_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J111429.54+072809.5_esdL0_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J230256.53+121310.2_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J144151.55+043738.5_sdL4_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J134852.93+101611.8_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J205721.89+005628.7_M7_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J225902.14+115602.1_sdL0_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J224749.77+053207.9_M7_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J024035.36+060629.3_sdM7_OSIRIS_GTC39-12B_Primeval-IV.fits",
                "ULAS_J134423.98+280603.8_sdL4_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J020858.62+020657.0_esdL3_OSIRIS_GTC63-13A_Primeval-III.fits",
                "ULAS_J004716.65+161242.4_M9_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J141832.35+025323.0_sdL0_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J225902.14+115602.1_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J011711.98-005213.4_M7_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J004602.85+091131.2_M9_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J011840.73+084424.7_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J124104.75-000531.4_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits"
                ]: low, high = 250, 1150
            
            elif filename in [
                "ULAS_J230443.30+093423.9_sdL0_OSIRIS_GTC63-13A_Primeval-IV.fits",
                "ULAS_J130710.22+151103.4_sdL8_OSIRIS_GTC46-14A_Primeval-IV.fits",
                "ULAS_J001931.33+063111.0_M9_OSIRIS_GTC39-12B_Primeval-IV.fits"
            ]: low, high = 300, 1000

            elif filename in [
                "ULAS_J141832.35+025323.0_sdL0_comb_OSIRIS_scombine_Primeval-IV.fits"
            ]: low,high = 300, 980

            elif filename in [
                "2MASS_J00412179+3547133_sdL0.5_OSIRIS_GTC80-15A_Primeval-I.fits"
            ]: low, high = 100-700

            # Masking
            if low is not None and high is not None:
                print(f"  Masking region {low}-{high}")

                # Apply mask 
                mask = np.logical_or(
                    spectrum.spectral_axis.value < low,
                    spectrum.spectral_axis.value > high
                )

                masked_spectrum = Spectrum(
                    spectral_axis=spectrum.spectral_axis,
                    flux=spectrum.flux,
                    mask=mask,
                    meta={'header': header.copy()}
                )
                # Plot masked spectrum 
                print("    Masked spectrum")
                masked_spectrum.plot()
                plt.show()
                file_plotted += 1

                # Add mask to header
                masked_spectrum.meta['header'].set("MASKED", True, "Spectrum has masked regions")
                header.set("MASK_LOW", low, comment="Lower bound of masked region")
                header.set("MASK_HIGH", high, comment="Upper bound of masked region")
            else:
                # Just check plottable and plot once
                if check_spectrum_plottable(spectrum, raise_error=True, show_plot=True):
                    print(f"  Plottable file: {filename}")
                    file_plotted += 1

            # Calculate TMID = MJD-OBS + (EXPTIME/2)/60*60*24 and add to header
            mjd_obs = header.get("MJD-OBS")
            exptime = header.get("EXPTIME")
            if mjd_obs is None or exptime is None:
                print(f"  WARNING: MJD-OBS or EXPTIME missing in header for {filename}. Cannot calculate TMID.")
            if mjd_obs is not None and exptime is not None:
                tmid = mjd_obs + (exptime / 2) / (60 * 60 * 24)
                header.set("TMID", tmid, "[d] MJD of exposure mid-point")
                print(f"  Added TMID to header: {tmid:.5f} d")


            # Save processed spectrum
            out_filename = filename.replace(".fits", "_SIMPLE.fits")
            out_filepath = os.path.join(output_path, out_filename)
            masked_spectrum.write(out_filepath, format="tabular-fits", overwrite=True)
            # fits.writeto(out_filepath, masked_spectrum.flux.value, header=header, overwrite=True)
            print(f"  Saved processed spectrum: {out_filename}")

        except Exception as e:
            print(f"  ERROR reading {filename}: {e}")
            file_failed += 1
            failed_files.append(filename)
            continue

    print(f"\nTotal files plotted: {file_plotted}")
    print(f"Total files failed: {file_failed}")
    if failed_files:
        print("Failed files: ", failed_files)


# Run
make_new_fitsfiles(path, output_path)
```
