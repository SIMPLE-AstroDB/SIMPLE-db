```python
from astrodb_utils.spectra import check_spectrum_plottable
from specutils.manipulation import median_smooth
from astropy.io import fits
import astropy.units as u
import os
import pandas as pd
import numpy as np
from specutils import Spectrum1D, Spectrum
from astropy.wcs import WCS
import matplotlib.pyplot as plt
```

```python
path = "/Users/guanying/SIMPLE db/SIMPLE-db/scripts/spectra_convert/zhang18/sty2054_supplemental_files/"
```

This notebook is to convert:  
-  13 Xshooter NIR spectra with smoothed and unsmoothed graph visualize

```python
# NIR Xshooter
file_plotted = 0
file_failed = 0
failed_files = []
for filename in os.listdir(path):


    if filename.endswith("fits") and "xshooter" in filename.lower() and "NIR" in filename.upper():

        file_path = os.path.join(path, filename)
        try:
            with fits.open(file_path) as hdul:
                print(f"\nReading FITS file: {filename}")
                header = hdul[0].header

                spectrum = Spectrum.read(file_path, format="wcs1d-fits", flux_unit="erg / (cm2 s Angstrom)")

                # Plot and check
                if check_spectrum_plottable(spectrum, raise_error=True, show_plot=True):
                    print(f"Plotable file name: {filename}")
                    file_plotted += 1

                # Try smoothing
                print("Trying smooth spectrum")
                smoothed_spectrum = median_smooth(spectrum, width=101)
                check_spectrum_plottable(smoothed_spectrum, raise_error=True, show_plot=True)

        except Exception as e:
            print(f"Error reading {filename}: {e}")
            print(f"Header info: {header}")
            file_failed += 1
            failed_files.append(filename)

print(f"\nTotal files plotted: {file_plotted}")
print(f"Total files failed: {file_failed}")
print("Failed files: ", failed_files)
```

```python
filename = "SDSS_J010448.46+153501.8_usdL1.5_Xshooter_NIR_Primeval-II.fits"
file_path = os.path.join(path, filename)

try:
    with fits.open(file_path) as hdul:
        print(f"Reading FITS file: {filename}")
        
        data = hdul[0].data
        header = hdul[0].header.copy()
        print("Flux unit:", header["BUNIT"])
        print("Data shape:", data.shape)        
        # Create 1D WCS 
        wcs = WCS(header, naxis=1)
        
        spectrum = Spectrum(
            flux=data * u.Unit(header["BUNIT"]),
            wcs=wcs
        )
        # Matplotlib plot
        plt.figure(figsize=(10, 5))
        plt.plot(spectrum.spectral_axis, spectrum.flux, lw=0.8)
        plt.xlabel(f"Wavelength ({spectrum.spectral_axis.unit})")
        plt.ylabel(f"Flux ({spectrum.flux.unit})")
        plt.title(filename)
        plt.grid(True)
        plt.show()
        
        # Smooth
        print("Smoothing spectrum")
        smoothed_spec = median_smooth(spectrum, width=101)
        check_spectrum_plottable(smoothed_spec, show_plot=True)

except Exception as e:
    print(f"Error reading {filename}: {e}")
```
