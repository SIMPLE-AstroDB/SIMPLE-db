```python
from specutils import Spectrum
import os
import matplotlib.pyplot as plt
from specutils import Spectrum
```

```python
"""
This notebook is to plot all new FITS file spectra created
"""
```

```python
fits_dir = "/Users/guanying/SIMPLE db/SIMPLE-db/scripts/spectra_convert/BONES Archive/Processed BONES/"
fits_file = os.listdir(fits_dir)
fits_file = [f for f in fits_file if f.endswith('.fits')]
plotted_spectra = 0
failed_spectra = 0
for file in fits_file:
    file_path = os.path.join(fits_dir, file)
    spectrum = Spectrum.read(file_path, format="tabular-fits")
    print(f"Checking {file}...")

    try:
        spectrum.plot()
        plt.show()
        print(f"plottable!\n")
        plotted_spectra += 1
    except Exception as e:
        print(f"   Error plotting {file}: {e}")
        failed_spectra += 1
        continue

print(f"Plotted spectra: {plotted_spectra}")
print(f"Failed spectra: {failed_spectra}")  

""" 
Plotted spectra: 66
Failed spectra: 0
"""
```
